#Redução de Dimensionalidade em Imagens para Redes Neurais

```python
from PIL import Image
```

```python
#montando o drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
imagem = Image.open('/content/drive/MyDrive/datasets/JillValentine.jpg')
display(imagem)
```

Os valores 0.2989, 0.5870 e 0.1140 são frequentemente usados como pesos porque refletem a proporção das sensibilidades relativas do olho humano às diferentes cores. Eles são derivados da recomendação CIE 1931, que é um padrão internacional para a representação das cores.

```python
largura, altura = imagem.size
imagem_cinza = Image.new('L', (largura, altura))

for x in range(largura):
  for y in range(altura):
    r, g, b = imagem.getpixel((x,y))
    cinzas = int(0.2989 * r + 0.5870 * g + 0.1140 * b)
    imagem_cinza.putpixel((x, y), cinzas)
display(imagem_cinza)
```

O limiar de cor para ser preto ou branco é decido em 128

```python

linha= 128
largura, altura = imagem_cinza.size
imagem_pb = Image.new('1',  (largura, altura))

for x in range(largura):
  for y in range(altura):
    pixel = imagem_cinza.getpixel((x,y))
    binario = 0 if pixel < linha else 255
    imagem_pb.putpixel((x,y), binario)
display(imagem_pb)
```
